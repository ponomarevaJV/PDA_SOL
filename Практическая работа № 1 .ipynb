### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(url)
drinks.head ()
```

### Step 4. Which continent drinks more beer on average?

```python
result = drinks.groupby('continent')['beer_servings'].max()
print(result)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
result = drinks.groupby('continent')['beer_servings'].describe()
print(result)
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks.groupby('continent').mean()
```

```python
df_1 = drinks.groupby('continent').mean()
df_1.to_csv('Ponomareva Julia Vladimirovna.csv', index=True)
from google.colab import files
files.download('Ponomareva Julia Vladimirovna.csv')
```

```python
drinks.groupby('continent').median()

```
